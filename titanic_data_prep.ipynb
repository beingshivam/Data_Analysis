```
import pandas as pd
```

```
titanic = pd.read_csv("train.csv", index_col=0)
titanic.head()
```

```
titanic.info()
```

```
#delete cabin col

titanic.drop('Cabin',axis=1,inplace=True)
```

```
titanic.head(5)
```

```
#number of males and females

titanic.groupby('Sex').size()
```

```
#renaming col value

titanic['Embarked'] = titanic.Embarked.map(lambda e: 'SI' if e=='S' else e)

```

```
titanic[['Embarked']].head(5)
```

```
#categorizing data with respect to age and adding it into new col

def f(age):
  if age<18:
    return "kid"
  elif age >=18:
    return "adult"
  else:
    return "not known"
  

titanic["Category"] = titanic.Age.map(f)
```

```
titanic.head()
```

```
titanic["Namelength"] = titanic.Name.str.len()
```

```
titanic.head()
```

```
#max length name
import numpy as np

max_name_length = np.max(titanic.Namelength)
titanic[titanic['Namelength'] == max_name_length]
```

```
#find missing age rows

titanic[pd.isnull(titanic.Age)].head()
```

```
#rows where age is present

age_data = titanic[pd.notnull(titanic.Age)]

age_data.head()
```

```
#grouping passengers based on age index

age_data.set_index('Age',inplace=True)
```

```
age_data.head()
```

```
#resetting index

age_data.reset_index(inplace=True)
```

```
age_data.head()
```

```
#mean of people between age 10-40

age_df = age_data[(age_data.Age>10) & (age_data.Age<40)]
```

```
age_df.Age.mean()
```

```
#filling missing age

titanic["Age"]= titanic.Age.fillna(26.18)
```

```
titanic.head()
```

```
titanic.info()
```

```
titanic.groupby("Survived").count()
```

```
gbs = titanic.groupby("Survived")

gbs.groups
```

```
def f(r):
  m= np.sum(r=='male')
  f= np.sum(r=='female')
  return('Male',m,'Female',f)


gbs.Sex.agg(f)
```
