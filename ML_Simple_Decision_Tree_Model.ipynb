```
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

```
home_file = "train.csv"

home_data = pd.read_csv(home_file, index_col=0)
home_data.head()
```

```
#prediction/target variable

Y = home_data.SalePrice
```

```
#choosing features

home_features = ["LotArea","YearBuilt","1stFlrSF","2ndFlrSF","FullBath","BedroomAbvGr","TotRmsAbvGrd"]

X = home_data[home_features]
```

```

X.describe()
```

```
X.head()
```

```
#building a model

home_model = DecisionTreeRegressor(random_state=1)

#fit model
home_model.fit(X,Y)

print("Predictions for 10 houses:")
print(X.head(10))
print("The predicted prices are:")
y_pred = home_model.predict(X.head(10))
print(y_pred)
```

```
#actual prices of first 10 houses

y_test = home_data["SalePrice"].head(10)
```

```
#evaluate accuracy
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```
